```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import random
```

```python
df=pd.read_csv('/content/default of credit card clients.csv')
```

```python
X=df.iloc[:,0:24]
y=df['default payment next month']
```

```python
n_samples = 10
test_size = 0.3
random_state = 42
accuracies = []
```

```python
bestAccuracy = 0
bestKernel = ""
best_accuracy = 0
best_gamma = 0
iteration = 1000
kernelList = ('rbf', 'poly', 'linear',  'sigmoid')
result_dict={}
```

```python
final_c=0
final_gamma=0
final_kernel=""
final_accuracy=0
```

```python
for key in range(n_samples):
    X=df.iloc[:,0:24]
    y=df['default payment next month']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=random_state)
    bestAccuracy = 0
    bestKernel = ""
    best_accuracy = 0
    best_gamma = 0  
    random_state += 1
    kernel = random.choice(kernelList)
    C = random.uniform(0,1)
    gamma = random.uniform(0,1)
    model = SVC(kernel=kernel,C=C,gamma=gamma,max_iter=1000)
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test,y_pred)
    if accuracy > bestAccuracy:
      bestAccuracy = accuracy
      bestKernel = kernel
      best_C = C
      best_gamma = gamma
    if final_accuracy<accuracy:
      final_gamma=best_gamma
      final_c=best_C
      final_kernel=bestKernel   
      final_accuracy=accuracy
    result_dict[key] = {'accuracy' : bestAccuracy, 'kernel' : bestKernel, 'C' : best_C, 'gamma' : best_gamma}
    print('Best accuracy for sample ',key,' is ',bestAccuracy)
```

```python
print(result_dict)
```

```python
result_df = pd.DataFrame(result_dict)
result_df = result_df.transpose()
result_df
```

```python
max_accuracy = result_df['accuracy'].max()
best_sample_key = result_df[result_df['accuracy'] == max_accuracy].index[0]
best_sample_key
```
